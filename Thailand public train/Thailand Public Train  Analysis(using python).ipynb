```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import calendar as cal
```

```python
df = pd.read_csv('drt2566_02.csv')
df.info()
```

```python
df
```

```python
df.head(15)
```

```python
df['Date'] = pd.to_datetime(df['Date'],dayfirst=False)
print(df['Date'].dtype)
```

```python
df['Weekday'] = df['Date'].dt.day_name()
df['Weekday_Num'] = df['Date'].dt.weekday
df['Month'] = df['Date'].dt.month_name()
df['Year'] = df['Date'].dt.year
```

```python
df
```

```python
StationEng = {'Station_Code':[f'A{i}' for i in range(1,9)],
              'StationEng':['Suvarnabhumi','Lad Krabang','Ban Thap Chang','Hua Mak','Ramkhamhaeng','Makkasan','Ratchaprarop','Phayathai']}
StationEng_df = pd.DataFrame(StationEng)
StationEng_df
```

```python
df = df.merge(StationEng_df,how='left',on='Station_Code')
WithOut2025 = df[df['Year'] < 2025]
WithOut2025
```

```python
#Percentage of Passenger Arrival
ArrPerStation = (WithOut2025.groupby(by='StationEng')['Passenger_Arrival'].sum()/WithOut2025['Passenger_Arrival'].sum()*100).sort_values(ascending=False)
ArrPerStation
```

```python
ArrPerStation_df = ArrPerStation.reset_index()
ArrPerStation_df.columns = ['Station_Name','Percentages']
x_pos = np.arange(len(ArrPerStation_df['Station_Name']))*1.25
fig,ax = plt.subplots()
RGB_code = [np.random.rand(3,) for _ in range(len(ArrPerStation_df['Station_Name']))]
ax.bar(x_pos,ArrPerStation_df['Percentages'],tick_label=ArrPerStation_df['Station_Name'],color=RGB_code)
ax.set_xticks(x_pos)
ax.set_xticklabels(ArrPerStation_df['Station_Name'],rotation=45,ha='center')
plt.xlabel('Station Name',labelpad=20)
plt.ylabel('Percentages',labelpad=10)
plt.title('Arrival By Station')
plt.show()
```

```python
#Percentage of Passenger Departure
DepPerStation = (WithOut2025.groupby(by='StationEng')['Passenger_Departure'].sum()/WithOut2025['Passenger_Departure'].sum()*100).sort_values(ascending=False)
DepPerStation
```

```python
DepPerStation_df = DepPerStation.reset_index()
DepPerStation_df.columns = ['Station_Name','Percentages']
x_pos = np.arange(len(DepPerStation_df['Station_Name']))*1.25
fig,ax = plt.subplots()
RGB_code = [np.random.rand(3,) for _ in range(len(DepPerStation_df['Station_Name']))]
ax.bar(x_pos,DepPerStation_df['Percentages'],tick_label=DepPerStation_df['Station_Name'],color=RGB_code)
ax.set_xticks(x_pos)
ax.set_xticklabels(DepPerStation_df['Station_Name'],rotation=45,ha='center')
plt.xlabel('Station Name',labelpad=20)
plt.ylabel('Percentages',labelpad=10)
plt.title('Departure By Station')
plt.show()
```

```python
ArrByWeek = (WithOut2025.groupby(by='Weekday')['Passenger_Arrival'].sum()/WithOut2025['Passenger_Departure'].sum()*100)
ArrByWeek
```

```python
ArrByWeek_df = ArrByWeek.reset_index()
ArrByWeek_df.columns = ['Weekday','Percentages']
ordered_days = list(cal.day_name)

# เรียงข้อมูลตามลำดับที่กำหนด
ArrByWeek_df = ArrByWeek_df.set_index('Weekday').loc[ordered_days].reset_index()

x_pos = np.arange(len(ArrByWeek_df['Weekday']))*1.25
fig,ax = plt.subplots(layout = 'constrained')
RGB_code = [np.random.rand(3,) for _ in range(len(ArrByWeek_df['Weekday']))]
# พล็อตกราฟ
r = ax.bar(x_pos,ArrByWeek_df['Percentages'],tick_label = ArrByWeek_df['Weekday'],color=RGB_code)
ax.bar_label(r,padding=3)
ax.set_xticks(x_pos)
plt.tight_layout()
plt.xlabel('WeekDay',labelpad=15)
plt.ylabel('Percentages',labelpad=15)
plt.ylim(0,18)
plt.title('Arrival Passenger By Weekday')
plt.show()
```

```python
DepByWeek = (WithOut2025.groupby(by='Weekday')['Passenger_Departure'].sum()/WithOut2025['Passenger_Departure'].sum()*100)
DepByWeek
```

```python
DepByWeek_df = DepByWeek.reset_index()
DepByWeek_df.columns = ['Weekday','Percentages']
ordered_days = list(cal.day_name)

# เรียงข้อมูลตามลำดับที่กำหนด
DepByWeek_df = DepByWeek_df.set_index('Weekday').loc[ordered_days].reset_index()

x_pos = np.arange(len(DepByWeek_df['Weekday']))*1.25
fig,ax = plt.subplots(layout = 'constrained')
RGB_code = [np.random.rand(3,) for _ in range(len(DepByWeek_df['Weekday']))]
# พล็อตกราฟ
r = ax.bar(x_pos,DepByWeek_df['Percentages'],tick_label = DepByWeek_df['Weekday'],color=RGB_code)
ax.bar_label(r,padding=3)
ax.set_xticks(x_pos)

plt.xlabel('WeekDay',labelpad=15)
plt.ylabel('Percentages',labelpad=15)
plt.ylim(0,18)
plt.title('Departure Passenger By Weekday')
plt.show()
```

```python


# ตัวอย่างจำลอง 
ArrByWeek_df_count = ArrByWeek_df
ArrByWeek_df_1 = pd.DataFrame({'Percentages': list(ArrByWeek_df['Percentages'])})
DepByWeek_df_1 = pd.DataFrame({'Percentages': list(DepByWeek_df['Percentages'])})

# แปลงคอลัมน์ Percentages เป็น Series
ArrPer = pd.Series(ArrByWeek_df_1['Percentages'])
DepPer = pd.Series(DepByWeek_df['Percentages'])

# แปลงเป็น numpy array สำหรับกราฟ
percentages = {
    'Arrival Count': ArrPer.to_numpy(),
    'Departure Count': DepPer.to_numpy(),
}

# ชื่อวันในสัปดาห์
weekdays = tuple(cal.day_name)

x = np.arange(len(weekdays))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

# วาดกราฟ
for attribute, measurement in percentages.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    multiplier += 1

# ตั้งค่ารายละเอียดกราฟ
ax.set_ylabel('Percentages')
ax.set_title('Arrival and Departure by Weekday')
ax.set_xticks(x + width)
ax.set_xticklabels(weekdays)
ax.legend(loc='upper left', ncols=2)
ax.set_ylim(0, 17)
ax.grid(True, axis='y', linestyle='--', alpha=0.6)

plt.show()
```

```python
ArrPerYear = (WithOut2025.groupby(by='Year')['Passenger_Arrival'].sum()/1000000)
ArrPerYear
```

```python
ArrPerYear_df = ArrPerYear.reset_index()
ArrPerYear_df.columns = ['Year','Passengers']
ArrPerYear_df['Year'] = ArrPerYear_df['Year'].astype(str)
plt.plot(ArrPerYear_df['Year'],ArrPerYear_df['Passengers'],marker='o')
for i,j in zip(ArrPerYear_df['Year'],ArrPerYear_df['Passengers']):
    plt.text(i,j+1,f'{j:.2f}',ha = 'center')
plt.title('Passengers Arrival Per Year')
plt.xlabel('Year')
plt.ylabel('Passengers(M)')
plt.ylim(10,40)

plt.show()
```
