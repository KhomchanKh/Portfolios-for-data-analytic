```python
import pandas as pd
import matplotlib.pyplot as mlt
import requests
import os
from dotenv import load_dotenv
import json
```

```python
load_dotenv()
api_key = os.getenv('MY_API_KEY')
```

```python
url = f"https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=NVDA&apikey={api_key}"
data = requests.get(url).json()
print(json.dumps(data,indent=2))
```

```python
stocks = ['NVDA','MSFT','AAPL','AMZN','GOOGL','META','TSLA']
all_df = []
for stock in stocks:
    url = f"https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol={stock}&apikey={api_key}"
    data_add = requests.get(url).json()    
    timeseries = data_add.get('Time Series (Daily)', {})
    df = pd.DataFrame.from_dict(timeseries, orient='index')
    df['Stock'] = stock
    df.index = pd.to_datetime(df.index)
    df.sort_index(inplace=True)
    all_df.append(df)

combined_df = pd.concat(all_df).reset_index().rename(columns={'index': 'Date',
                                                              '1. open':'Open',
                                                              '2. high':'High',
                                                              '3. low':'Low',
                                                              '4. close':'Close',
                                                              '5. volume':'Volume'})
combined_df
```
